```python
from bs4 import BeautifulSoup
import requests
from splinter import Browser
import pandas as pd
```

```python
url="https://mars.nasa.gov/news/?page=0&per_page=40&order=publish_date+desc%2Ccreated_at+desc&search=&category=19%2C165%2C184%2C204&blank_scope=Latest"
```

```python
response= requests.get(url)
```

```python
soup = BeautifulSoup(response.text, 'html.parser')
```

```python
print(soup.prettify())
```

```python
title_rows = soup.find_all('div', class_="content_title")
```

```python
# Loop through returned results
titles=[]

for title_row in title_rows:
    
    # Identify and return titles
    title = title_row.find('a').text
    title=title.strip('\n')

    # Print results only if title, price, and link are available
    titles.append(title)
    
```

```python
titles
```

```python
previews = soup.find_all('div', class_="rollover_description_inner")
```

```python
descriptions=[]

for preview in previews:
    description= preview.text
    description=description.strip('\n')
    descriptions.append(description)
    
descriptions
```

```python
url3= "https://astrogeology.usgs.gov/search/results?q=hemisphere+enhanced&k1=target&v1=Mars"
response2=requests.get(url3)
```

```python
soup3 = BeautifulSoup(response2.text, 'html.parser')
soup3
```

```python
soup3= soup3.find_all("div", class_="item")
soup3
```

```python
sp3=soup3[0].img
sp3=str(sp3)
sp3=sp3.split("=")
sp3=sp3[3].strip('/>')
sp3=sp3.strip('"')



sp3
```

```python
image_urls=[]
for x in range(len(soup3)):
    img=soup3[x].img
    img=str(img)
    img=img.split("=")
    img=img[3].strip('/>')
    img=img.strip('"')
    image_url = f'https://astrogeology.usgs.gov{img}'
    image_urls.append(image_url)
    
image_urls
```
